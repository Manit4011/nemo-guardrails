```python
def analyze_conversations(df):
    user_queries = df[df['role'] == 'user']['query']
    
    unique_questions = user_queries.nunique()
    most_frequent_questions = Counter(user_queries).most_common(5)
    
    print(f"Total unique questions: {unique_questions}")
    print("Top 5 most asked questions:")
    for question, count in most_frequent_questions:
        print(f"Question: {question} | Count: {count}")
```
